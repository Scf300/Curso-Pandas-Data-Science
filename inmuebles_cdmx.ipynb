# **Demandas ML**

 **Cuáles son los valores promedio de alquiler por tipo de propiedad?**

```python
datos.head()
```

```python
datos["Valor"].mean()
```

```python
# valores promedio de alquiler
datos.groupby("Tipo")["Valor"].mean(numeric_only= True)
```

```python
datos.groupby("Tipo")["Valor"].mean(numeric_only=True).sort_values()
```

```python
# valores promedio de alquiler por tipo de propiedad
df_tipo_precio= datos.groupby("Tipo")["Valor"].mean(numeric_only=True).sort_values()
df_tipo_precio.plot(kind="barh", figsize= (8,8), color ="purple")
```
