# **Demandas ML**

 **Cuáles son los valores promedio de alquiler por tipo de propiedad?**

```python
datos.head()
```

```python
datos["Valor"].mean()
```

```python
# valores promedio de alquiler
datos.groupby("Tipo")["Valor"].mean(numeric_only= True)
```

```python
datos.groupby("Tipo")["Valor"].mean(numeric_only=True).sort_values()
```

```python
# valores promedio de alquiler por tipo de propiedad
df_tipo_precio= datos.groupby("Tipo")["Valor"].mean(numeric_only=True).sort_values()
df_tipo_precio.plot(kind="barh", figsize= (8,8), color ="purple")
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Casa Comercial','Terreno Estándar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio', 'Industria']
```

```python
datos.query("@inmuebles_comerciales in Tipo")
```

```python
datos.query("@inmuebles_comerciales not in Tipo")
```

```python
df = datos.query("@inmuebles_comerciales not in Tipo")
df.head()
```

```python
df_tipo_precio= df.groupby("Tipo")["Valor"].mean(numeric_only=True).sort_values()
df_tipo_precio.plot(kind="barh", figsize= (8,8), color ="purple")
```

**¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?"**

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts()
```

```python
df.Tipo.value_counts(normalize=True)
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame()
porcentaje_tipo.plot(kind="bar", figsize= (12,8), color ="green", xlabel="Porcentajes", ylabel="Porcentaje")
```

```python
df.query("Tipo == 'Departamento'")
```

```python
df = df.query("Tipo == 'Departamento'")
df.head(2)
```

```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
# Guardando el DataFrame en una variable
df_ejemplo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')

```

```python
# Cambiando el nombre de la columna "Tipo" a "Porcentajes"
df_ejemplo.rename(columns={'Tipo': 'Porcentajes'}, inplace=True)

# Visualizando el DataFrame
df_ejemplo
```

**Verificar cuántas colonias únicas existen en nuestra base de datos.**

```python
datos["Colonia"].unique()
```

**Analizar qué colonias tienen el promedio de alquiler más alto.**

```python
datos.groupby("Colonia")["Valor"].mean(numeric_only= True)
```

```python
colonias_alquiler_alto = datos.groupby("Colonia")["Valor"].mean(numeric_only= True).sort_values(ascending=False)
colonias_alquiler_alto
```

**Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.**

```python
df_cinco_colonias_mas_altos = colonias_alquiler_alto.head(5)
```

```python
colonias_alquiler_alto.head(5)
```

```python
df_cinco_colonias_mas_altos.plot(kind="bar", figsize= (12,8), color ="green", xlabel="Colonias", ylabel="Alquiler")
```
